```python
from keras.models import load_model
import pandas as pd
from google.colab import drive
from keras.preprocessing.sequence import pad_sequences
from keras.preprocessing.text import Tokenizer
```

```python
drive.mount('/content/drive')
data=pd.read_csv('/content/drive/MyDrive/BigDataAnalystic/PROJECT/Prediction_Dataset.csv')
```

```python
# Loading the models
lstm = load_model('/content/drive/MyDrive/BigDataAnalystic/PROJECT/LSTM_Model.h5')
print("LSTM Model loaded.")
gru = load_model('/content/drive/MyDrive/BigDataAnalystic/PROJECT/GRU_Model.h5')
print("GRU Model loaded.")
```

```python
Text=data['Review'].astype(str)
```

```python
text=Text.iloc[100]
text
```

```python
tokenizer = Tokenizer()
text_seq = tokenizer.texts_to_sequences([text])
max_len = len(text_seq[0])
text_pad = pad_sequences(text_seq, maxlen=max_len)
```

```python
# Check if the text_pad variable is empty
if text_pad.size == 0:
    # Generate the padded text sequences
    tokenizer = Tokenizer()
    text_seq = tokenizer.texts_to_sequences([text])
    max_len = len(text_seq[0])
    text_pad = pad_sequences(text_seq, maxlen=max_len)

# Check if the text_pad variable has the correct shape
if text_pad.shape[1] != 1104:
    # Adjust the maxlen parameter of the pad_sequences function
    max_len = 1104
    text_pad = pad_sequences(text_seq, maxlen=max_len)
```

```python
# Make the prediction
Prediction1 = lstm.predict(text_pad)
Prediction1 = (Prediction1 > 0.5).astype(int).flatten()
print("LSTM Prediction:", "positive" if Prediction1[0] == 1 else "negative")
```

```python
# Make the prediction
Prediction2 = gru.predict(text_pad)
Prediction2 = (Prediction2 > 0.5).astype(int).flatten()
print("GRU Prediction:", "positive" if Prediction1[0] == 1 else "negative")
```
